# Virat Kolhi ODI Innings EDA

## Libraries Import

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import warnings
warnings.filterwarnings("ignore")
```

## Data Load

```python
df = pd.read_csv("virat_kohli_odi_innings_data.csv")
```

```python
df.info()
```

```python
df.head(3)
```

## Data Cleaning

Runs Column Data Cleaning

```python
df['Runs'].unique()
```

```python
df['Runs'].replace("TDNB","0",inplace=True)
df['Runs'].replace("DNB","0",inplace=True)
df['Runs'] = df['Runs'].str.replace("*", "", regex=False)
```

```python
df["Runs"]=df["Runs"].astype('int64')
```

Mins Column Data Cleaning

```python
df['Mins'].unique()
```

```python
df['Mins'].replace("-","0",inplace=True)
df["Mins"]=df["Mins"].astype('int64')
```

BF Column Data Cleaning

```python
df['BF'].unique()
```

```python
df['BF'].replace("-","0",inplace=True)
df["BF"]=df["BF"].astype('int64')
```

4s and 6s Column Data Cleaning

```python
df['4s'].unique()
```

```python
df['4s'].replace("-","0",inplace=True)
df['4s']=df['4s'].astype('int64')
```

```python
df['6s'].unique()
```

```python
df['6s'].replace("-","0",inplace=True)
df['6s']=df['6s'].astype('int64')
```

SR Column Data Cleaning

```python
df['SR'].unique()
```

```python
df['SR'].replace("-","0",inplace=True)
df['SR']=df['SR'].astype('float64')
```

Pos Column Data Cleaning

```python
df['Pos'].unique()
```

```python
df['Pos'].replace("-","0",inplace=True)
df['Pos']=df['Pos'].astype('int64')
```

Dismissal Column Data Cleaning

```python
df['Dismissal'].unique()
```

```python
df['Dismissal'].replace("0","lbw",inplace=True)
```

Ground Column Data Cleaning

```python
df['Ground'].unique()
```

Opposition Column Data Cleaning

```python
df['Opposition'].unique()
```

Inns Column Data Cleaning

```python
df['Inns'].unique()
```

```python
df['Inns'].replace("-","1",inplace=True)
df['Inns']=df['Inns'].astype('int64')
```

## Graphs

### What are the top 5 grounds where Virat Kohli scored the most runs

```python
ground_runs = df.groupby('Ground')['Runs'].sum().sort_values(ascending=False).head(5)
```

```python
plt.figure(figsize=(10,6))
ground_runs.plot(kind='bar', color='c')
plt.title('Top 5 Grounds by Total Runs')
plt.xlabel('Ground')
plt.ylabel('Total Runs')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Against which opposition does Virat Kohli have the highest average score.

```python
opposition_avg_runs = df.groupby('Opposition')['Runs'].mean().sort_values(ascending=False)
```

```python
plt.figure(figsize=(10,6))
opposition_avg_runs.plot(kind='bar', color='cyan')
plt.title('Average Runs Against Different Oppositions')
plt.xlabel('Opposition')
plt.ylabel('Average Runs')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### What is the distribution of his scores (Runs) in ODIs.

```python
plt.figure(figsize=(10,6))
plt.hist(df['Runs'], bins=20, color='cyan', edgecolor='black')
plt.title('Distribution of Runs')
plt.xlabel('Runs')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### What is Virat Kohli's strike rate distribution across different innings (first vs. second innings).

```python
plt.figure(figsize=(10,6))
df.boxplot(column='SR', by='Inns', grid=False, patch_artist=True, boxprops=dict(facecolor='cyan'))
plt.title('Strike Rate Distribution in Different Innings')
plt.suptitle('')
plt.xlabel('Innings (1st or 2nd)')
plt.ylabel('Strike Rate')
plt.tight_layout()
plt.show()
```

### How does his scoring pattern vary based on the number of boundaries (4s and 6s).

```python
df['Boundaries'] = df['4s'] + df['6s']
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['Boundaries'], df['Runs'], color='cyan')
plt.title('Runs vs Boundaries')
plt.xlabel('Total Boundaries (4s + 6s)')
plt.ylabel('Runs')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### What is the distribution of the number of balls faced in an inning

```python
plt.figure(figsize=(10,6))
plt.hist(df['BF'], bins=20, color='cyan', edgecolor='black')
plt.title('Distribution of Balls Faced')
plt.xlabel('Balls Faced (BF)')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()
```

 ### How does his batting position affect his runs scored

```python
plt.figure(figsize=(10,6))
df.boxplot(column='Runs', by='Pos', grid=False, patch_artist=True, boxprops=dict(facecolor='cyan'))
plt.title('Runs by Batting Position')
plt.suptitle('')
plt.xlabel('Batting Position')
plt.ylabel('Runs')
plt.tight_layout()
plt.show()
```

### What are the most frequent types of dismissals for Virat Kohli

```python
dismissal_counts = df['Dismissal'].value_counts()
```

```python
plt.figure(figsize=(10,6))
dismissal_counts.plot(kind='bar', color='cyan')
plt.title('Types of Dismissals for Virat Kohli')
plt.xlabel('Dismissal Type')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Which venues have seen his best performances in terms of centuries and half-centuries

```python
half_centuries = df[(df['Runs'] >= 50) & (df['Runs'] < 100)]
centuries = df[df['Runs'] >= 100]

ground_50s = half_centuries['Ground'].value_counts().head(5)
ground_100s = centuries['Ground'].value_counts().head(5)
```

```python
plt.figure(figsize=(10,6))
plt.bar(ground_50s.index, ground_50s.values, label='50s', color='lightblue')
plt.bar(ground_100s.index, ground_100s.values, bottom=ground_50s.values, label='100s', color='lightgreen')
plt.title('Best Venues for 50s and 100s')
plt.xlabel('Ground')
plt.ylabel('Number of 50s and 100s')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
